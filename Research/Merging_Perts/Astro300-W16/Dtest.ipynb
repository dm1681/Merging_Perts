## 30 Doradus (NGC 2070)

```python
ngc2070 = FixedTarget.from_name("ngc2070")
obj = ngc2070
print (obj.ra.hms)

#Date chosen is Dec. 14
date_obj_highest = Time("2016-12-14 00:00:00", format='iso')

obs = sso

astro_set = obs.twilight_evening_astronomical(date_obj_highest, which='nearest')  
astro_rise = obs.twilight_morning_astronomical(date_obj_highest, which='next')

start_time = astro_set
end_time = astro_rise
delta_t = end_time - start_time
observe_time = start_time + delta_t * np.linspace(0.0, 1.0, 50)

plot_sky(obj, obs, observe_time);
```

```python
# Run this for airmass of ngc2070
plot_airmass(obj, obs, observe_time)
```
